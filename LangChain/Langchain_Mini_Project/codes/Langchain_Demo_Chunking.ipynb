```python
import os

file_path = "C:/Users/Pooja Nayak/Desktop/Yodaplus/LangChain/Langchain_Mini_Project/codes/pdfs/All_Course_Slides.pdf"

if os.path.exists(file_path):
    print("PDF mil gaya bhai!")
else:
    print("PDF nahi mila ðŸ˜¢")
```

```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load PDF
loader = PyPDFLoader(file_path) 
pages = loader.load()

# Split into chunks
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
documents = splitter.split_documents(pages)
```

```python
from langchain.embeddings import SentenceTransformerEmbeddings

embedding_model = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")  # Free & good model
```

```python
from langchain.vectorstores import Milvus

milvus_connection_args = {
    "host": "localhost",
    "port": "19530"
}

vectorstore = Milvus.from_documents(
    documents,
    embedding=embedding_model,
    connection_args=milvus_connection_args,
    collection_name="my_collection"
)
```

```python
query = "yodaplus" # Write your Question here.

results = vectorstore.similarity_search(query, k=15)
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA
from dotenv import load_dotenv

load_dotenv() 
openai_key = os.getenv("OPENAI_API_KEY")

# LangChain LLM setup
llm = ChatOpenAI(openai_api_key=openai_key, model_name="gpt-3.5-turbo")

# Retrieval QA chain setup
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vectorstore.as_retriever(search_kwargs={"k": 15}),
    return_source_documents=True
)

response = qa_chain({"query": query})
```

```python
print("ðŸ”¹ Question:", query)
print("\nðŸ¤– Answer:\n", response["result"])

print("\nðŸ“š Source Chunks:")
for i, doc in enumerate(response["source_documents"]):
    print(f"\n--- Chunk {i+1} ---\n{doc.page_content[:300]}...\n")
```
